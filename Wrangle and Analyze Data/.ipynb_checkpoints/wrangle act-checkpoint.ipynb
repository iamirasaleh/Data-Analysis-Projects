```python
#putting them into 1 file
id_tweet =[]
favorite_count=[]
retweet_count =[]

with open ('tweet-json.txt', 'r') as file:
    for line in file:
        I = re.findall(r' "id": ([0-9]+),' ,  line)
        F = re.findall(r' "favorite_count": ([0-9]+),' ,  line)
        R = re.findall(r' "retweet_count": ([0-9]+),' ,  line)  
          
        if I:
            id_tweet.append(I[0])
        if F:
            favorite_count.append(F[0])
        if R:
            retweet_count.append(R[0]) 

#print the final lists
print(id_tweet[0])
print(favorite_count[0])
print(retweet_count[0])
```

```python
# Create a dictionary with the data
data = {'id_tweet': id_tweet, 'favorite_count': favorite_count, 'retweet_count': retweet_count}

# Create a dataframe from the dictionary
twitter_data = pd.DataFrame(data)

# Print the dataframe
#print(df3.shape)
```

### Visualization

### Insight #1: is there any relationship between the favourite and retweet counts? 

```python
# Plotting a scatterplo
colors = ["#c0abff"] 
sns.set_palette(sns.color_palette(colors))
sns.lmplot(x = "favorite_count", 
           y = "retweet_count", 
           data = twitter_archive_master,
           height = 5,
           aspect = 1.6,
           scatter_kws = {'alpha':1/3, "color": "#4b3296"})  #643B9F
plt.title('Favourite and Retweet');
plt.xlabel('Favourite Count');
plt.ylabel('Retweet Count');
```

**as shown in the plot, there is a relationship between favourite and retweet counts. 
and actually it is a Strong Positive relationship, which means the most twitter user likes these types of tweet;
the most users would retweet it**

### Insight #2: Rating Counts

```python
# Get Rating Counts
colors = ["#917ad6"]
com_rate = twitter_archive_master.groupby('rating_numerator').tweet_id.count().sort_values(ascending = True)
com_rate.plot(kind = 'barh', figsize = [13, 9])

for index, value in enumerate(com_rate):
    plt.text(value, index, float(value))
    
plt.title('The Most rating', fontsize = 18)
plt.xlabel('Tweet Count', fontsize = 17)
plt.ylabel('Rating', fontsize = 17)
plt.xlim(0, 650);
```

**in this Bar graph, we can say the increasing of tweet count leads to more rating in tweets.**

### Insight #3: Source of the most retweet

```python
color2 = ['#4b3296', '#917ad6', '#8b77c7', '#ab9cd9']
sns.set_palette(sns.color_palette(color2))
sou = sns.countplot(data = df, x = 'source')
sou.figure.suptitle('Tweet Sources', fontsize = 15);
plt.xticks(rotation = 90); 
plt.xlabel('source', fontsize = 14);
plt.ylabel('Rating', fontsize = 14);
```
